```python
import pandas as pd

transactions = pd.read_excel("retail_sales_dataset.xlsx", sheet_name="Transactions")
products = pd.read_excel("retail_sales_dataset.xlsx", sheet_name="Products")

merged = transactions.merge(products, on="ProductID")

top_products = merged.groupby("ProductName")["Quantity"].sum().sort_values(ascending=False)

print(top_products)
```

```python
transactions = pd.read_excel("retail_sales_dataset.xlsx", sheet_name="Transactions")
products = pd.read_excel("retail_sales_dataset.xlsx", sheet_name="Products")

merged = transactions.merge(products, on="ProductID")

merged["Revenue"] = merged["Quantity"] * merged["UnitPrice"]
merged["Profit"] = (merged["UnitPrice"] - merged["CostPrice"]) * merged["Quantity"]

print(merged.head())
```

```python
top_profit = merged.groupby("ProductName")["Profit"].sum().sort_values(ascending=False)

print(top_profit.head(10))
```

```python
top_revenue = merged.groupby("ProductName")["Revenue"].sum().sort_values(ascending=False)

print(top_revenue.head(10))
```
