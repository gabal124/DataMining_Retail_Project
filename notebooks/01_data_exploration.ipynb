```python
import pandas as pd

# 1. Load the Dataset
file_path = "../online_retail_II.csv"
df = pd.read_csv(file_path)

# 2. Dataset Preview (First 5 rows)
print("--- Dataset Preview ---")
print(df.head())

# 3. Dataset Size & Information
print("\n--- Dataset Info (Size & Data Types) ---")
print(df.info())

# 4. Checking for Quality Issues
print("\n--- Data Quality Issues ---")

# A. Missing Values
print("\nMissing Values per Column:")
print(df.isnull().sum())

# B. Duplicate Records
print(f"\nTotal Duplicate Rows: {df.duplicated().sum()}")

# C. Statistics (To detect negative values in Quantity/UnitPrice)
print("\nBasic Statistics (Check for Negative Values):")
print(df[['Quantity', 'Price']].describe())

# D. Check for unique customers
print(f"\nUnique Customers: {df['Customer ID'].nunique()}") 
```
